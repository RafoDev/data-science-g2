```python
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = "https://randomuser.me/api/?results=100"
response = requests.get(url)
data = response.json()
users = data["results"]
users[0]
```

```python
genders = [user["gender"] for user in users]
ages = [user["dob"]["age"] for user in users]
countries = [user["location"]["country"] for user in users]
print(genders)
print(ages)
print(countries)
```

```python
df_users = pd.DataFrame({
    "gender": genders,
    "age": ages,
    "country": countries
})
df_users
```

```python
sns.countplot(data=df_users, y="country", hue="gender")
plt.show()
```

```python
sns.countplot(data=df_users, x="country", hue="gender")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.histplot(data=df_users, x="age", kde=True)
plt.show()
```

```python
sns.scatterplot(data=df_users, x = df_users.index, y="age", hue="gender")
plt.xlabel("Index")
plt.show()
```
