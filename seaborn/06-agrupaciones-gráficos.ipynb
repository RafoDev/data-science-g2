```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
tips.head(2)
```

Joint Plot

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="hist", hue="sex")
plt.show()
```

Pair Plot

```python
sns.pairplot(data=tips, corner=True, hue="sex")
plt.show()
```
