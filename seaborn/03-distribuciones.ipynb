```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
tips.head(5)
```

Histplot

```python
# stat: Define que estadística calcula por cada bin
# count (frecuencia), density(conteo ajustado bajo 1), probability (ajustado como probabilidad)
# multiple: layer(default), stack, dodge
sns.histplot(
    data=tips,
    x="tip",
    hue="sex",
    cumulative=False,
    stat="count",
    multiple="dodge")

plt.show()
```

```python
sns.kdeplot(data=tips, x="tip", hue="sex", fill=True, bw_adjust=.5)
plt.show()
```
