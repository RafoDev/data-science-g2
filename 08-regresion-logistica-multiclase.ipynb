# Undersampling

```python
sns.countplot(df["Class"])
plt.xticks(rotation=45)
plt.show()
```

```python
from imblearn.under_sampling import RandomUnderSampler
undersample = RandomUnderSampler(random_state=42)
```

```python
X = df.drop("Class", axis=1)
y = df["Class"]
```

```python
X_under, y_under = undersample.fit_resample(X, y)
```

```python
sns.countplot(y_under)
plt.xticks(rotation=45)
plt.show()
```

# Codificación Numérica `Class`

```python
classes = list(np.unique(y_under))
```

```python
y_under.replace(classes,[1,2,3,4,5,6,7],inplace=True)
```

# Entrenamiento del modelo

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_under, y_under, random_state=42, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score

model = LogisticRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
print(confusion_matrix(y_test, predictions))
print(accuracy_score(y_test, predictions))
```

# Ajuste de Hiperparámetros
**Hyperparams Tunning**

```python
def logistic_model(solver_, multiclass_):
  logistic_regression_model = LogisticRegression(random_state=42, solver=solver_, multi_class=multiclass_, n_jobs=-1)
  return logistic_regression_model

model = logistic_model("saga", "multinomial")
model.fit(X_train, y_train)
predictions = model.predict(X_test)
print(confusion_matrix(y_test, predictions))
print(accuracy_score(y_test, predictions))
```

```python
multiclass = ["ovr", "multinomial"]
solver_list = ["newton-cg", "lbfgs", "sag", "saga"]
scores = []
params = []

for i in multiclass:
  for j in solver_list:
    try:
      model = logistic_model(j, i)
      model.fit(X_train, y_train)
      predictions = model.predict(X_test)

      params.append(i + '-' + j)
      acc = accuracy_score(y_test, predictions)
      scores.append(acc)
    except:
      None
```

```python
sns.barplot(x=scores, y=params).set_title("CLFs accuracies")
```

```python
model = logistic_model("newton-cg", "ovr")

model.fit(X_train, y_train)

predictions = model.predict(X_test)

print(confusion_matrix(y_test, predictions))
print(accuracy_score(y_test, predictions))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

conf_matrix = confusion_matrix(y_test, predictions)

display_conf_matrix = ConfusionMatrixDisplay(confusion_matrix=conf_matrix)

display_conf_matrix.plot(cmap="gray")
```
