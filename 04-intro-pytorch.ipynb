# Primera NN con PyTorch

```python
import torch.nn as nn
import torch.optim as optim
```

**Arquitectura básica de la NN**

```python
# La clase hereda de nn.Module, que es la clase base para todas las NNs de Pytorch
class SimpleNN(nn.Module):
  def __init__(self):
    super(SimpleNN, self).__init__()
    self.fc = nn.Linear(2, 1) # ó nn.Linear(in_features=2, out_features=1)

  def forward(self, x): # Define como fluyen los datos a través de la red
    return self.fc(x)
```

**Instanciar el modelo**

```python
model = SimpleNN()
model
```

**Función de pérdida y Optimizador**


```python
criterion = nn.MSELoss() # ERROR CUADRÁTICO MEDIO - mide que tanto se equivocó el modelo
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

**Datos de Prueba**

```python
inputs = torch.tensor([[1.0, 2.0],[2.0, 3.0]], dtype=torch.float32)
label = torch.tensor([[1.0],[2.0]], dtype=torch.float32)
inputs.shape, label.shape
```

**Entrenamiento**

```python
for epoch in range(10): # cada época es una pasada completa por los datos de entrenamiento
  output = model(inputs) # el modelo genera predicciones

  loss = criterion(output, label) # compara las predicciones con los valores reales

  # backpropagation
  optimizer.zero_grad() # borra los actuales
  loss.backward() # calcula nuevos pesos
  optimizer.step() # actualiza los pesos

  print(f"epoch {epoch + 1}, loss: {loss.item():.4f}")
```
