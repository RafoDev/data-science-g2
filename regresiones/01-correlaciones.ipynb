```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,100,100)
y = x**2
```

```python
sns.scatterplot(x=x,y=y)
plt.show()
```

# Coeficiente de correlación de Pearson

```python
np.corrcoef(x,y)
```

```python
penguins_df = sns.load_dataset("penguins")
penguins_df.head()
```

```python
penguins_df.corr(numeric_only=True)
```

```python
sns.heatmap(
    data=penguins_df.corr(numeric_only=True),
    vmin=-1,
    vmax=1,
    center=0,
    annot=True,
    cmap="BrBG",
    linewidth=0.5
)
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.clustermap(
    data=penguins_df.corr(numeric_only=True),
    vmin=-1,
    vmax=1,
    center=0,
    annot=True,
    cmap="BrBG",
    linewidth=0.5
)
plt.show()
```

```python
(
    penguins_df
    .corr(numeric_only=True)[["body_mass_g"]]
    .sort_values(by="body_mass_g", ascending=False)
    .pipe(
        lambda df:
          sns.heatmap(
          data=df,
          vmin=-1,
          vmax=1,
          center=0,
          annot=True,
          cmap="BrBG",
          linewidth=0.5
      )
    )
)
plt.show()
```

```python
sns.scatterplot(
    data=penguins_df,
    x="flipper_length_mm",
    y="body_mass_g"
)
plt.show()
```

```python
sns.scatterplot(
    data=penguins_df,
    x="bill_length_mm",
    y="bill_depth_mm"
)
plt.show()
```
