Dos conclusiones opuestas sobre los mismos datos.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style(style="whitegrid")
```

```python
penguins_df = sns.load_dataset("penguins")
penguins_df.dropna(inplace=True)
penguins_df.head()
```

```python
sns.heatmap(data=penguins_df.corr(numeric_only=True), cmap="BrBG", vmin=-1, vmax=1, annot=True)
plt.show()
```

## Analizamos la relación entre: bill_length y bill_depth

```python
# Mientras más largo el pico, menos profundo.

sns.regplot(
    data=penguins_df,
    x="bill_length_mm",
    y="bill_depth_mm"
)
plt.show()
```

```python
# Mientras más largo el pico, más profundo.

sns.lmplot(
    data=penguins_df,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species"
)
plt.show()
```

```python
sns.lmplot(
    data=penguins_df,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species"
)
sns.regplot(
    data=penguins_df,
    x="bill_length_mm",
    y="bill_depth_mm"
)
plt.show()
```

```python
sns.pairplot(data=penguins_df, hue="species")
plt.show()
```
