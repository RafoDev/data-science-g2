# Limitaciones de la regresión lineal simple

1. La regresión lineal de X-Y no es igual a Y-X: No hay simetría.
2. Correlación no implica causalidad.
3. Solo se pueden representar relaciones lineales.

Comprobamos la simetría

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy
```

```python
penguins_df = sns.load_dataset("penguins")
penguins_df.dropna(inplace=True)
penguins_df.head()
```

## Regresión lineal de x a y

* x = bill_length
* y = bill_depth

---

* x = bill_depth
* y = bill_length


```python
x = penguins_df.bill_length_mm
y = penguins_df.bill_depth_mm

reg_x_y = scipy.stats.linregress(x=x, y=y)
reg_y_x = scipy.stats.linregress(x=y, y=x)

print(reg_x_y)
print(reg_y_x)
```

```python
sns.scatterplot(x=x, y=y)
fx_1 = np.array([x.min(), x.max()])
fy_1 = reg_x_y.slope * fx_1 + reg_x_y.intercept
plt.plot(fx_1, fy_1, "r")
plt.show()
```

```python
sns.scatterplot(x=y, y=x)
fx_2 = np.array([y.min(), y.max()])
# y = m * x + b
fy_2 = reg_y_x.slope * fx_2 + reg_y_x.intercept
plt.plot(fx_2, fy_2, "g")
plt.show()
```

```python
sns.scatterplot(x=x, y=y)
plt.plot(fx_1, fy_1, "r")
plt.plot(fy_2, fx_2, "g")
plt.show()
```
