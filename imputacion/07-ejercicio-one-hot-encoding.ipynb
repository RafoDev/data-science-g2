```python
import janitor
import pandas as pd
import sklearn.compose
import sklearn.preprocessing
```

```python
data_df = pd.read_csv("/content/data_ecommerce_customer_churn.csv")
data_df.head()
```

```python
data_df = (
    data_df.clean_names(case_type="snake")
)
data_df
```

```python
transformer = sklearn.compose.make_column_transformer(
    (sklearn.preprocessing.OrdinalEncoder(), ["marital_status"]),
    (sklearn.preprocessing.OneHotEncoder(), ["prefered_order_cat"]),
    remainder="passthrough"
)
transformer
```

```python
data_transformed_df = (
    pd.DataFrame(
        transformer.fit_transform(data_df),
        columns = transformer.get_feature_names_out(),
        index = data_df.index
    )
    .rename_columns(
        function = lambda x: x.removeprefix("ordinalencoder__")
    )
    .rename_columns(
        function = lambda x: x.removeprefix("onehotencoder__")
    )
    .rename_columns(
        function = lambda x: x.removeprefix("remainder__")
    )
)
data_transformed_df
```
