```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Cambiando el estilo a los gr치ficos

```python
sns.set_style(style="whitegrid")
penguin_color={
    "Adelie":"#ff6602ff",
    "Chinstrap":"#0f7175ff",
    "Gentoo":"#c65dc9ff",
}
```

## Cargar el dataframe

```python
penguins_df = sns.load_dataset("penguins")
penguins_df.head()
```

# Conteo de variables

```python
penguins_df.describe(include="all")
```

```python
penguins_df.describe(include=[np.number])
```

# Categorizar

```python
penguins_df.dtypes
```

```python
penguins_df = penguins_df.astype({
    "species":"category",
    "island":"category",
    "sex":"category",
})
```

```python
penguins_df.describe(include=["category"])
```

# An치lisis de variables

**Graficar el conteo de specie**

```python
(penguins_df
    .species
    .value_counts())
```

```python
(
    penguins_df
    .species
    .value_counts()
    .plot(
        kind="bar",
        color=penguin_color.values(),
    )
)
plt.show()
```

```python
(
    penguins_df
    .species
    .value_counts()
    .plot(
        kind="pie",
        color=penguin_color.values(),
        autopct="%1.1f%%"
    )
)
plt.show()
```

```python
penguins_df["variable"] = ""
penguins_df
```

```python
(
    penguins_df
    .pipe(
        lambda df:(
            sns.displot(
                data=df,
                x="variable",
                hue="species",
                multiple="fill",
                palette=penguin_color
            )
        )
    )
)
plt.show()
```
