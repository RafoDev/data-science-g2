# Graficamos la media, mediana y moda de una variable

```python
sns.histplot(
    data=penguins_df,
    x="bill_depth_mm"
)

bill_mean = round(penguins_df.bill_depth_mm.mean(), 2)
bill_median = penguins_df.bill_depth_mm.median()
bill_mode = penguins_df.bill_depth_mm.mode()[0]

#print(bill_mean, bill_median, bill_mode)

plt.axvline(
    x=bill_mean,
    color="red",
    label=f"Media: {bill_mean}",
    linestyle="dashed"
)

plt.axvline(
    x=bill_median,
    color="green",
    label=f"Mediana: {bill_median}",
    linestyle="dashed"
)

plt.axvline(
    x=bill_mode,
    color="yellow",
    label=f"Moda: {bill_mode}",
    linestyle="dashed"
)

plt.legend()
plt.show()
```
