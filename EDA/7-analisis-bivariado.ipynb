```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.set_style(style="whitegrid")
```

```python
penguins_df = sns.load_dataset("penguins")
penguins_df.head()
```

# Relación entre 2 variables

```python
sns.scatterplot(data=penguins_df, x="bill_length_mm", y="bill_depth_mm", alpha=0.5, s=50)
plt.show()
```

```python
sns.displot(data=penguins_df, x="bill_length_mm", y="bill_depth_mm")
plt.show()
```

```python
sns.displot(data=penguins_df, x="bill_length_mm", y="bill_depth_mm", kind="kde")
plt.show()
```

```python
sns.jointplot(data=penguins_df, x="flipper_length_mm", y="body_mass_g", marginal_kws=dict(bins=25, fill=False))
plt.show()
```

## Relación entre variables numéricas y categóricas

```python
sns.scatterplot(data=penguins_df, x="species", y="flipper_length_mm", hue="species", alpha=0.5, s=50)
plt.show()
```

```python
#plt.figure(figsize=(8,4))
penguin_color ={
    'Adelie': '#edbe66',
    'Gentoo': '#0f7175ff',
    'Chinstrap':'#c65dc9ff'
}
sns.stripplot(data=penguins_df, x="species", y="flipper_length_mm",palette=penguin_color,s=3)
plt.show()
```

```python
sns.boxplot(data=penguins_df, y="species", x="flipper_length_mm", palette=penguin_color)
sns.stripplot(data=penguins_df, y="species", x="flipper_length_mm", color="0.3", s=3)
plt.show()
```

```python
sns.violinplot(data=penguins_df, x="species", y="flipper_length_mm", color="0.8")
sns.stripplot(data=penguins_df, x="species", y="flipper_length_mm", palette=penguin_color, s=3)
plt.show()
```
