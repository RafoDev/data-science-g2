# Entrenamiento del modelo de regresión logística (clasificación)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=42) # opcional shuffle=False
```

```python
from sklearn.linear_model import LogisticRegression
logistic_reg = LogisticRegression()
logistic_reg.fit(X_train, y_train)
```

# Evaluación del modelo

```python
predictions = logistic_reg.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_test, predictions)
conf_matrix
```

```python
sns.heatmap(conf_matrix, cmap="coolwarm", annot=True)
```
