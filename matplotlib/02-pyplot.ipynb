```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = x ** 2
y
```

plot

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, color="r")
plt.show()
```

```python
plt.plot(x,y, color="r", marker="o")
plt.show()
```

```python
plt.plot(x, y, color="r", marker="o", linestyle=":")
plt.show()
```

FMT : format

```python
plt.plot(x, y, "ro:")
plt.show()
```

```python
plt.plot(x, y, "ro:", y, x, "g*-")
plt.show()
```

plt.show

```python
plt.plot([2,3,5,3,2])
plt.plot([1,3,6,2,1])
plt.show()
```

hist

```python
plt.hist(x)
plt.show()
```

bar

```python
plt.bar(x,y)
plt.show()
```

```python
plt.pie(x)
plt.show()
```
