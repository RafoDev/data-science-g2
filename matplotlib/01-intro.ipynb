# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y1 = np.random.rand(10).cumsum()
y2 = np.random.rand(10).cumsum()
```

```python
print(y1,y2)
```

* **Figura**: Espacio que abarcan las "gráficas" (color celeste)
* **Conjunto de Ejes**: Lineas ortogonales que definene las dimensiones
* **Gráfica**: Lo que se encuentra dentro de cada conjunto de ejes

```python
# FIGURA > CONJUNTO/S DE EJES > GRÁFICA/S
# RAIZ > TRONCO > RAMAS

"""
Explicación:
  Una "Figura" puede tener uno o varios "Conjuntos de Ejes"
  y un "Conjunto de Ejes" puede tener varios "Graficos"
"""

fig, ax = plt.subplots(nrows = 1, ncols = 2)
fig.set_facecolor("LightBlue")
fig.set_size_inches(8,4)
ax[0].plot(y1)
ax[1].hist(y2)
plt.show()
```

## Formas de crear un mismo gráfico

```python
plt.plot(y1)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(y1)
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(y1)
plt.show()
```

## Interfaces de programación

Estilo Matlab

```python
plt.figure()

plt.subplot(1,2,1)
plt.plot([2,6,7,5,2])
plt.title("Ventas")
plt.xticks(range(0,5), ["Ene", "Feb", "Mar", "Abr", "May"])

plt.subplot(1,2,2)
plt.title("Almacenes")
plt.hist(y2, bins=5)

plt.show()
```

Interfaz Orientada a Objetos

```python
fig, ax = plt.subplots(1,2)
ax[0].plot([2,6,7,5,2])
ax[0].set_title("Ventas")
ax[0].set_xticks(range(0,5), ["Ene", "Feb", "Mar", "Abr", "May"])

ax[1].hist(y2, bins=5)
ax[1].set_title("Almacenes")

plt.show()
```
