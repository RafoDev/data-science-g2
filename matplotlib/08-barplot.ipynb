```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
paises = ["PERU", "COLOMBIA", "MEXICO", "BOLIVIA", "CHILE"]
poblacion = [100,800,900,100,300]

plt.bar(paises, poblacion, color=["r", "b", "g"])
plt.xticks(np.arange(5), ("Perú", "Colombia", "Mexico", "Bolivia", "Chile"), rotation=45)
plt.show()
```

```python
plt.barh(paises, poblacion)
plt.show()
```
