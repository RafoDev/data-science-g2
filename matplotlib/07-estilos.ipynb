```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,5,11)
```

Estilos

```python
print(plt.style.available)
```

```python
plt.style.use("seaborn-v0_8-bright")
```

```python
fig, axes = plt.subplots()
axes.set_title("Gráfico 1")
axes.plot(x, x+1, 'r--')
plt.show()
```

```python
#linestyle, marker
fig, axes = plt.subplots()
axes.plot(x, x+1, color="#e66712")
axes.plot(x, x+2, color="#12e695", marker=".", linestyle="dashdot")
axes.plot(x, x+3, color="#5c12e6", marker="8")
# -
# -
# -
plt.show()
```
