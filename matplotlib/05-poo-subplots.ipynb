```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,"b")
plt.show()
```

```python
fig, axes = plt.subplots(nrows = 1, ncols= 2)
axes[0].plot(x,y,"b--")
axes[1].hist(y)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.plot(x,y,"b--")
ax2.hist(y)
plt.show()
```

```python
fig, axes = plt.subplots(2,4)
axes[0,0].plot(x, np.cos(x))
axes[0,1].plot(x, np.sin(x), "y")
axes[0,2].plot(x, np.tan(x), "r")
axes[0,3].plot(x, np.cos(x) ** 2)
axes[1,0].plot(x, np.cos(y))
axes[1,1].plot(x, np.sin(y), "g")
axes[1,2].plot(x, np.tan(y), "y")
axes[1,3].plot(x, np.cos(y) ** 2)
fig.tight_layout()
plt.show()
```
