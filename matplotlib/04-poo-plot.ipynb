```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,5,11)
y = x ** 2
```

Plot con POO

rect -> left, bottom, width, height

```python
fig = plt.figure(figsize=(8,4))
axes = fig.add_axes([0.1, 0.1, 0.5, 0.9])
axes.plot(x, y ,"b")
fig.show()
```

```python
fig = plt.figure(figsize=(8,4))
axes = plt.axes([0.1, 0.1, 0.5, 0.9])
axes.plot(x, y ,"b")
fig.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.15, 0.5, 0.4, 0.3])

axes1.plot(x,y,"b")
axes2.plot(y,x,"r")
axes2.set_facecolor("gray")

fig.show()
```
