```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs
X, y, c = make_blobs(500, cluster_std=0.8, centers=4, n_features=2, return_centers=True)
print(X) #Muestras
print(y) #Etiquetas
print(c) #Centroides
```

# Creamos DataFrames con valores sint√©ticos


```python
df_blobs = pd.DataFrame(
    {
        'x1': X[:,0],
        'x2': X[:,1],
        'y': y
    }
)
df_blobs
```

```python
df_centers = pd.DataFrame(
    {
        'x1': c[:,0],
        'x2': c[:,1],
    }
)
df_centers
```

# Visualizamos los datos

```python
sns.scatterplot(data=df_blobs, x='x1', y='x2')
sns.scatterplot(data=df_centers,x='x1', y='x2', marker='X', s=100)
plt.show()
```

# Aplicamos K-means

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)
df_cluster = kmeans.fit_predict(X)
df_cluster
```

```python
df_blobs['cluster'] = df_cluster
df_blobs
```

```python
k_means_centers = kmeans.cluster_centers_
df_k_means_centers = pd.DataFrame(
    {
        'x1': k_means_centers[:,0],
        'x2': k_means_centers[:,1],
    }
)
df_k_means_centers
```

# Visualizamos resultados del clustering con K-Means

```python
sns.scatterplot(data=df_blobs, x='x1', y='x2', hue="cluster", palette="coolwarm")
sns.scatterplot(data=df_centers, x='x1', y='x2', marker="X", s=150, color="green")
sns.scatterplot(data=df_k_means_centers, x='x1', y='x2', marker="o", s=150, color="purple")
plt.show()
```
