Filtro por condiciones

```python
filter_2016 = df_books["Year"] > 2016
filter_2016
```

```python
df_books[filter_2016]
```

```python
filter_genre = df_books["Genre"] == "Fiction"
df_books[filter_genre]
```

```python
df_books[filter_2016 & filter_genre]
```

```python
df_books[filter_2016][filter_genre]
```

```python
# ~: es una negación:
# True -> False
# False -> True
df_books[~filter_2016]
df_books[-filter_2016]
```
