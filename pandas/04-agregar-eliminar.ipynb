Dataframe.drop

df.drop(labels, axis, inplace)

* labels: Etiqueta/s de filas o columnas que se quieren eliminar
* axis: 0 -> Filas, 1 -> Columnas
* inplace: True -> modifica el original, False -> devuelve una copia con los cambios
