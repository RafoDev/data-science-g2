```python
import pandas as pd
```

**Series**

Una serie de pandas es como una columna de una hoja de Excel.
Es una estructura de datos similiar a una lista de Python, pero más poderosa (numpy).

```python
pd.Series([10,20,30,40])
```

```python
paises = pd.Series(["Perú", "Bolivia", "Chile", "Colombia"], index=[1, 3, 6, 10])
paises
```

```python
paises.index
```

```python
dict1 = {1: "Perú", 3:"Bolivia", 6:"Chile", 10:"Colombia"}
pd.Series(dict1)
```

```python
paises[0:3]
```

**DataFrame**

Estructura de datos en forma de tabla, similar a una tabla de una base de datos. Es un conjunto de filas y columnas.

```python
dict_paises = {
    "paises": ["Perú", "Bolivia", "Chile", "Colombia"],
    "pbi":[183, 170, 160, 140],
    "habitantes_millones":[33,22,22,30]
}

df_paises = pd.DataFrame(dict_paises)
df_paises
```

```python
dict_paises = [{"pais": "Perú", "pbi":183}, {"pais": "Bolivia", "pbi":183}]
df_test = pd.DataFrame(dict_paises)
df_test
```
