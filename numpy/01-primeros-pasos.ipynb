## Gráfico Estadístico

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x,y)
plt.title("Grafico de Ejemplo")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

## Gráfico de Barras

```python
categorias = ["A", "B", "C", "D"]
valores = [5,7,3,8]

plt.bar(categorias, valores, color="green")
plt.title("Gráfico de barras")
plt.xlabel("Categorías")
plt.ylabel("Valores")
plt.show()
```
