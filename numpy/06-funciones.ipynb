Min y Argmin

```python
arr.min()
```

```python
# Obtiene el índice del elemento menor
# https://numpy.org/doc/stable/reference/generated/numpy.argmin.html
arr.argmin()
```

```python
matriz.min(0)
```

```python
matriz
```

```python
# Comparación entre los elementos de la D1
# Compara estos elementos:
# 0: [ 7, 10,  3, 16,  5]
# 1: [ 4, 15, 19,  8, 14]
#    [ 1,  0,  0,  1, 0]

matriz.argmin(0)
```

```python
# Comparación entre los elementos de la D2
# [ 7, 10,  3, 16,  5]
#           *
#   0   1   2   3   4 => 2
# [ 4, 15, 19,  8, 14]
#   *
#   0   1   2   3   4 => 0

matriz.argmin(1)
```

**PTP** (Peak to Peak)

La resta entre el máximo y el mínimo.
https://numpy.org/doc/2.0/reference/generated/numpy.ptp.html


```python
arr
```

```python
print(f"max: {arr.max()} - min: {arr.min()} = {arr.ptp()}")
```

```python
arr.ptp()
```
