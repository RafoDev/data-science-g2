```python
import pandas as pd
```

## Creamos un dataset de prueba

```python
df = pd.DataFrame.from_dict(
    data = {
        "a": list("abcdefghij"),
        "b": range(0,10)
    }
)
df
```

## Generamos nulos

```python
df.iloc[2:5, 0] = None
df.iloc[6:7, 1] = None
df
```

## Verificamos nulos

```python
df.isnull()
```

```python
df.isnull().sum().sum()
```

```python
df.size - df.isnull().sum().sum()
```

```python
df.isnull().sum()
```

Conteo por casos

```python
df.isnull().pipe(
    lambda df_1 : (
        df_1.sum()
        .reset_index(name="n_missing")
        .rename(columns={"index":"variable"})
        .assign(
            n_cases = len(df_1),
            pct_missing = lambda df_2: df_2.n_missing / df_2.n_cases * 100
        )
    )
)
```

## Extendiendo la API de pandas

```python
@pd.api.extensions.register_dataframe_accessor("missing")
class MissingMethods:
  def __init__(self, pandas_obj):
    self._df = pandas_obj

  def number_missing(self):
    return self._df.isnull().sum().sum()
```

```python
# [pandas]
# [pandas + (missing)]
# [pandas + (missing) + (correlations)]

df = pd.DataFrame.from_dict(
    data = {
        "a": list("abcdefghij"),
        "b": range(0,10)
    }
)
df.iloc[2:5, 0] = None
df.iloc[6:7, 1] = None
```

```python
df.missing.number_missing()
```
