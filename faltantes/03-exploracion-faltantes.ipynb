## Cargamos el dataset NHANES

```python
nhanes_df = (nhanes.load.load_NHANES_data(year="2017-2018"))
nhanes_df.head()
```

```python
nhanes_df = nhanes_df[[
    "GeneralHealthCondition",
    "Gender",
    "AgeStoppedBreastfeedingdays",
    "AgeStoppedReceivingFormuladays",
    "AgeStartedOtherFoodbeverage",
    "AgeFirstFedMilkdays"
]]
nhanes_df
```

# Exploración de datos faltantes

## Analisis con missing

### Resumen general

**Total de datos**

```python
nhanes_df.size
```

**Total de faltantes**

```python
nhanes_df.missing.number_missing()
```

**Total de completos**

```python
nhanes_df.missing.number_complete()
```

### Tabulaciones de datos faltantes

```python
nhanes_df.missing.missing_variable_summary().sort_values(by="pct_missing", ascending=False)
```

```python
nhanes_df.missing.missing_variable_table()
```

```python
nhanes_df.missing.missing_case_summary()
```

```python
nhanes_df.missing.sort_variables_by_missingness()
```
